```python
def create_dummies(df,colname):
    col_dummies = pd.get_dummies(df[colname],prefix=colname)
    col_dummies.drop(col_dummies.columns[0],axis=1,inplace=True)
    df = pd.concat([df,col_dummies],axis=1)
    df.drop(colname,axis=1,inplace=True)
    return df
```

```python
for c_feature in cat_data[['job','marital','education','contact','month','day_of_week','poutcome']]:
    cat_data[c_feature] = cat_data[c_feature].astype('category')
    cat_data = create_dummies(cat_data,c_feature)
```

```python
feature_cols = total_df.columns.difference(['y'])
X = total_df[feature_cols]
y = total_df['y']
```

# Feature Selection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
RFC = RandomForestClassifier()
RFC.fit(X,y)
score = np.round(RFC.feature_importances_,3)
```

```python
important = pd.DataFrame({'features':X.columns,'importance':score})
important = important.sort_values('importance',ascending= False).set_index('features')
plt.rcParams['figure.figsize'] = 11,4
important.plot.bar();
```

```python
rfe = RFE(RFC,n_features_to_select=15)
rfe = rfe.fit(X,y)
```

```python
rfe.get_support()
```

```python
import itertools
feature_maps = [(i,v) for i , v in itertools.izip_longest(rfe.get_support(),X.columns)]
```

```python
selected_features = [v for i,v in feature_maps if i== True]
print(selected_features)
```

```python
cols = X.columns[rfe.get_support()]
print(cols)
```

```python
X = total_df[cols]
y=total_df['y']
X.head()
```

# Model Training

```python
from sklearn.cross_validation import train_test_split
train_X,test_X,train_y,test_y = train_test_split(X,y,random_state = 123,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
LR = LogisticRegression()
LR.fit(train_X,train_y)
```

```python
y_pred = LR.predict(test_X)
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(LR.score(test_X, test_y)))
```

```python
scores = cross_val_score(LR,train_X,train_y,scoring='accuracy',cv=10)
```

```python
print("10-fold cross validation average accuracy: %.3f" % (scores.mean()))
```

```python
CM = confusion_matrix(test_y,y_pred)
print(CM)
print(classification_report(test_y,y_pred))
```

```python
NB_clf = GaussianNB()
NB_clf.fit(train_X,train_y)
```

```python
NB_Scores = cross_val_score(NB_clf,train_X,train_y,cv=10,scoring='accuracy')
print(NB_Scores.mean())
```

```python
print('Accuracy of naives Bayes classifier on test set: {:.2f}'.format(metrics.accuracy_score(test_y,NB_clf.predict(test_X))))
```

```python
CM = confusion_matrix(test_y,NB_clf.predict(test_X))
print(CM)
```

```python
KNN_clf = KNeighborsClassifier()
KNN_clf.fit(train_X,train_y)
```

```python
KNN_Scores = cross_val_score(KNN_clf,train_X,train_y,cv=10,scoring='accuracy')
print(KNN_Scores.mean())
```

```python
print('Accuracy of KNN classifier on test set: {:.2f}'.format(metrics.accuracy_score(test_y,KNN_clf.predict(test_X))))
```

```python
CM = confusion_matrix(test_y,KNN_clf.predict(test_X))
print(CM)
```
